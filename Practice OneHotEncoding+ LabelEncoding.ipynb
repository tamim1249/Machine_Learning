What is OneHotEncoder?
OneHotEncoder is used to convert categorical text data into numbers for machine learning.
It creates new columns for each category and uses 0/1 to indicate their presence — just like pd.get_dummies().

But OneHotEncoder is:

✅ More flexible
✅ Works nicely in ML Pipelines
✅ Handles unknown test categories
✅ Can return sparse matrix (memory efficient)

```python
import pandas as pd

df = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male'],
    'City': ['Dhaka', 'Khulna', 'Dhaka', 'Chittagong']
})
df
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df)
```

```python
#❗Output hobe sparse matrix (not directly visible). Dekhte chaile:
print(encoded.toarray())  # Convert to dense matrix
```

```python
#See Column Names
encoder.get_feature_names_out()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Sample DataFrame
df = pd.DataFrame({
    'Department': ['Sales', 'HR', 'IT', 'Sales', 'IT'],
    'Education': ['Bachelors', 'Masters', 'Bachelors', 'PhD', 'Masters'],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male'],
    'Experience': [3, 7, 2, 5, 10],
    'Salary': [40000, 50000, 42000, 60000, 75000]
})

# Step 1: Select categorical columns only
cat_cols = ['Department', 'Education', 'Gender']

# Step 2: Create encoder and fit_transform
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_data = encoder.fit_transform(df[cat_cols])

# Step 3: Get column names
column_names = encoder.get_feature_names_out(cat_cols)

# Step 4: Create encoded DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=column_names)

# Step 5: Combine with numeric columns
final_df = pd.concat([encoded_df, df[['Experience', 'Salary']]], axis=1)

# Step 6: Show result
print(final_df)
```

What is LabelEncoder?
LabelEncoder just converts categories → numbers.
But unlike OneHotEncoder, eta single column e integers boshae.
LabelEncoder ekta system ja category gulo ke number e convert kore.
LabelEncoder sudhu ekta column (1D array) er upor kaj kore.

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df = pd.DataFrame({
    'Department': ['Sales', 'IT', 'HR', 'IT', 'Sales', 'HR','l']
})

le = LabelEncoder()
df['Department_encoded'] = le.fit_transform(df['Department'])
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green'],
    'Size': ['Small', 'Large', 'Medium']
})

le = LabelEncoder()

for col in df.columns:
    df[col] = le.fit_transform(df[col])

print(df)
```

```python
from sklearn.preprocessing import OrdinalEncoder

df = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green'],
    'Size': ['Small', 'Large', 'Medium']
})

enc = OrdinalEncoder()
df_encoded = enc.fit_transform(df)
print(df_encoded)
```
