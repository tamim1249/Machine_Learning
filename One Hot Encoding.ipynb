pandas এর get_dummies() 

```python
import pandas as pd
df=pd.read_csv('homeprices.csv')
df
```

```python
d=pd.get_dummies(df.town)
d
```

```python
merged=pd.concat([df,d],axis='columns')
merged
```

```python
final=merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
X=final.drop(['price'],axis='columns')
X
```

```python
y=final.price
y
```

```python
model.fit(X,y)
model.predict([[2800,0,1]])
```

```python
model.score(X,y)
```

```python
model.predict([[3600,1,0]])
```

import pandas as pd
df=pd.read_csv('homeprices.csv')
d=pd.get_dummies(df.town)
merged=pd.concat([df,d],axis='columns')
final=merged.drop(['town','west windsor'],axis='columns')
from sklearn.linear_model import LinearRegression
model=LinearRegression()
X=final.drop(['price'],axis='columns')
y=final.price
model.fit(X,y)
model.predict([[2800,0,1]])

খানে মূলত pandas এর get_dummies() ফাংশন দিয়ে dummy variables তৈরি করা হয়েছে, যেটা One Hot Encoding এর মতোই কাজ করে।



Using sklearn OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dfle=df
dfle
```

```python
dfle.town=le.fit_transform(dfle.town) #town নামের categorical column কে numeric form-এ convert করছো:এটা হল Label Encoding, যা machine এর বোঝার উপযোগী করে categorical ডেটা কে রূপান্তর করে।




dfle
```

```python
X=dfle[['town','area']].values
X
```

```python
y=dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')

X=ct.fit_transform(X)
X
```

```python
X=X[:,1:] ## Dummy Variable Trap avoid করার জন্য 1st column বাদ দেওয়া হয়েছে
X
```

```python
model.fit(X,y)
model.predict([[1,0,2800]])
```
